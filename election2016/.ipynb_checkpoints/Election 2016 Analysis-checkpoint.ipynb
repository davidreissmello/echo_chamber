```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
echo_scores = pd.read_csv('echo_computation2016.csv')
```

```python
follower_information = pd.read_csv('../../All_Location_Gender.csv', sep='\t', header=1)
```

```python
echo = pd.merge(echo_scores, follower_information, left_on='followers', right_on='UserID')
echo['followers1_that_follow_candidate_not_opponent'] = echo.followers1_that_follow_candidate - echo.followers1_that_follow_candidate_opponent
echo['echo_chamber_only_candidate'] = echo.followers1_that_follow_candidate_not_opponent / echo.candidate_follower_followers1
echo['followers1_that_follow_opponent_not_candidate'] = echo.followers1_that_follow_opponent - echo.followers1_that_follow_candidate_opponent
echo['echo_chamber_only_opponent'] = echo.followers1_that_follow_opponent_not_candidate / echo.candidate_follower_followers1

echo.head()
```

```python
print("How many followers do we have information on:", echo.shape[0] / echo_scores.shape[0])
```

```python
def ends(df, x=5):
    """Returns both head and tail of the dataframe or series.
    
    Args:
        x (int): Optional number of rows to return for each head and tail
    """
    print('{} rows x {} columns'.format(np.shape(df)[0],np.shape(df)[1]))
    return df.head(x).append(df.tail(x))
```

```python
def candidate_classifier(x):
    if x == 1339835893:
        return 'Clinton'
    return 'Trump'
echo['candidate'] = echo.candidates.apply(candidate_classifier)
```

# By Candidate

```python
echo.groupby('candidate').agg({'echo_chamber':'mean',
                              'echo_chamber_only_candidate':'mean', 
                              'echo_chamber_only_opponent':'mean'})
```

# By State and Gender

```python
state_gender_echo_chamber_long = echo.groupby(['US-State', 'GenderGuesser0.4']).echo_chamber.mean()
state_gender_echo_chamber_long = pd.DataFrame(state_gender_echo_chamber_long)
state_gender_echo_chamber_long.reset_index(inplace=True)
state_gender_echo_chamber_wide = state_gender_echo_chamber_long.pivot(index='US-State', 
                                                                      columns='GenderGuesser0.4', 
                                                                      values='echo_chamber')

state_gender_echo_chamber_wide['gender_divide'] = np.abs(state_gender_echo_chamber_wide.female - state_gender_echo_chamber_wide.male)

ends(state_gender_echo_chamber_wide.sort_values('gender_divide'))
```

```python
state_gender_echo_chamber_long = echo.groupby(['US-State', 'GenderGuesser0.4']).echo_chamber_only_candidate.mean()
state_gender_echo_chamber_long = pd.DataFrame(state_gender_echo_chamber_long)
state_gender_echo_chamber_long.reset_index(inplace=True)
state_gender_echo_chamber_wide = state_gender_echo_chamber_long.pivot(index='US-State', 
                                                                      columns='GenderGuesser0.4', 
                                                                      values='echo_chamber_only_candidate')

state_gender_echo_chamber_wide['gender_divide'] = np.abs(state_gender_echo_chamber_wide.female - state_gender_echo_chamber_wide.male)

ends(state_gender_echo_chamber_wide.sort_values('gender_divide'))
```

```python
state_gender_echo_chamber_long = echo.groupby(['US-State', 'GenderGuesser0.4']).echo_chamber_only_opponent.mean()
state_gender_echo_chamber_long = pd.DataFrame(state_gender_echo_chamber_long)
state_gender_echo_chamber_long.reset_index(inplace=True)
state_gender_echo_chamber_wide = state_gender_echo_chamber_long.pivot(index='US-State', 
                                                                      columns='GenderGuesser0.4', 
                                                                      values='echo_chamber_only_opponent')

state_gender_echo_chamber_wide['gender_divide'] = np.abs(state_gender_echo_chamber_wide.female - state_gender_echo_chamber_wide.male)

ends(state_gender_echo_chamber_wide.sort_values('gender_divide'))
```

# By State by Candidate

```python
state_candidate_echo_chamber_long = echo.groupby(['US-State', 'candidate']).echo_chamber_only_candidate.mean()
state_candidate_echo_chamber_long = pd.DataFrame(state_candidate_echo_chamber_long)
state_candidate_echo_chamber_long.reset_index(inplace=True)
ends(state_candidate_echo_chamber_long.sort_values('echo_chamber_only_candidate'), 10)
```
